# Introduction to Pandas


We are now ready to take a closer look at the *data structures* provided by the [Pandas](https://pandas.pydata.org/docs/index.html) library. NumPy and its `ndarray` object serve as a basis for this topic (see the last lecture for more information).

### What is Pandas?

Pandas is built on top of NumPy and allows us to work efficiently with `DataFrames`. We can think of `DataFrames` as multidimensional arrays with attached `row` and `column` labels that often contain heterogeneous types of data (may include *missing data*). Pandas offers a number of powerful data operations that are also found in spreadsheets and databases.

Although the `ndarray` object already provides us with several ways to perform numerical computations, it clearly has some limitations when we try to use it for other tasks, such as performing operations that do not map well to element-wise broadcasting (e.g. [grouping](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html) and [pivoting](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.pivot.html)). `DataFrames`, `Series` and `Index`, which are the three main data structures in Pandas and, as mentioned above, are built on top of NumPy arrays, offer efficient ways to manipulate data.

In this lecture, we will explore the main and related data structures of Pandas (e.g. `DataFrames`, `Series` and `Index`, among others).

## Pandas Objects

We can think of Pandas objects as upgraded versions of NumPy arrays. The main difference is that in Pandas objects the `rows` and `columns` are identified with labels instead of simple integer indices as in NumPy arrays.

### Series Object



````python
import numpy as np
import pandas as pd
````


`Series` are one-dimensional arrays of indexed data. The general way to create a Pandas `Series` is `pd.Series(data, index=index)`, where `index` is an optional argument. The `data` argument can be various Python objects such as list or np array, scalar, or dictionary. We can create a `Series` from a list or array as follows:



````python
# Create a Pandas Series with default integer index
series_data = [1.5, 4.3, 27.5, 8, 1.7, 9, 10, 2.4]
series_1 = pd.Series(series_data)
print(series_1)
````


A `Series` object has both a sequence of values and a sequence of indices. These can be accessed with the `values` and `index` attributes, respectively. 

The `values` attribute returns a NumPy array:



````python
series_1.values # Get the underlying NumPy array
````


The `index` attribute returns a Pandas `Index` object, which is an array-like object that stores the labels for each element in the Series:



````python
series_1.index # Get the index object
````


We can access individual elements or slices of the `Series` using array-like indexing with square brackets, similar to NumPy arrays or Python lists:



````python
# Get the third value of series_1 (0-indexed)
print("Third element:", series_1[2])
# Get a slice from the first to the fourth value (exclusive of the last index)
print("Slice (0:4):"), series_1[0:4]
````


#### Series as a Generalized NumPy Array

`Series` are more general and flexible than standard NumPy arrays because they allow explicitly defined, custom indices. While NumPy arrays have an implicitly defined integer index, a Pandas `Series` can have an index consisting of values of any desired type (e.g., strings, dates, or even other numbers that are not sequential).

For example, we can create a `Series` with a string-based index:



````python
# Create a Pandas Series with a custom string index
series_data_custom_index = [1.5, 4.3, 27.5, 8, 1.7, 9, 10, 2.4]
custom_index = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
series_2 = pd.Series(series_data_custom_index, index=custom_index)
print(series_2)
````


Now, we can access values using these custom labels:



````python
series_2['h'] # Fetch the element in series_2 associated with index 'h'
````


#### Series as a Specialized Dictionary

`Series` objects share similarities with Python dictionaries, where arbitrary keys map to a set of values. This analogy becomes clearer when creating a `Series` from a dictionary.

Let's create a `Series` using a dictionary where keys represent labels and values represent data:



````python
data_dict = {'Zone1': 106633468.00, 'Zone2': 100775278.00, 'Zone3': 96878519.00,
             'Zone4': 169177603.00, 'Zone5': 29109577.00, 'Zone6': 55291468.00,
             'Zone7': 2924768.00}
swedish_auto_zone_pay = pd.Series(data_dict)
swedish_auto_zone_pay
````


In this case, the `Series` index is automatically derived from the dictionary's keys. We can fetch values using dictionary-style indexing:



````python
swedish_auto_zone_pay['Zone3']
````


However, `Series` also support array-type operations like slicing, which dictionaries do not:



````python
swedish_auto_zone_pay['Zone3':'Zone7'] # Slicing by explicit label, inclusive of 'Zone7'
````


> **💡 TIP**
> When an explicit `index` is provided from a dictionary, only the keys present in the `index` will be included in the `Series`. Missing keys will result in `NaN` values.
>



````python
data = {2:'Gerber', 3:'Albrecher', 1:'Hieber', 5:'Arnold', 4:'Dufresne'}
print(pd.Series(data, index = [1, 5, 4]))
print(pd.Series(data, index = [1,"cats", "dogs"]))
````


### DataFrame Object

`DataFrames` are also fundamental structures in Pandas.

#### DataFrame as a Generalized NumPy Array

Just as a `Series` is an upgraded one-dimensional NumPy array, a `DataFrame` can be thought of as an upgraded two-dimensional NumPy array. It features flexible `row` indices and `column` names, making it much more powerful for tabular data. A `DataFrame` can also be viewed as a collection of "aligned" `Series` objects, meaning they all share the same index.

Let's create an example using two `Series` objects for payments and claims in different zones:



````python
# Re-using swedish_auto_zone_pay from previous example
# Create a new Series for claims data
swedish_auto_zone_claims_dict = {'Zone1': 23174.00, 'Zone2': 21302.00, 'Zone3': 19938.00,
                                 'Zone4': 31913.00, 'Zone5': 5962.00, 'Zone6': 10262.00,
                                 'Zone7': 620.00}
swedish_auto_zone_claims = pd.Series(swedish_auto_zone_claims_dict)
print("Payments Series:\n", swedish_auto_zone_pay)
print("\nClaims Series:\n", swedish_auto_zone_claims)
````


Now, we combine these two `Series` into a single `DataFrame`:



````python
swedish_auto_zone_data = pd.DataFrame({'payments': swedish_auto_zone_pay, 'claims': swedish_auto_zone_claims})
print(swedish_auto_zone_data)
````


> **💡 TIP**
> In a Jupyter Notebook environment, `DataFrame` objects are typically displayed as user-friendly HTML tables. Like `Series`, `DataFrame` objects have `index` and `columns` attributes to access row and column labels:
>



````python
print("DataFrame Index:", swedish_auto_zone_data.index)
print("DataFrame Columns:", swedish_auto_zone_data.columns)
````


> **💡 TIP**
> A `DataFrame` is essentially a two-dimensional NumPy array where both rows and columns have explicit labels, providing a highly flexible structure for tabular data.
>

#### DataFrame as a Specialized Dictionary

A `DataFrame` can also be thought of as a dictionary where keys are column names and values are `Series` objects representing the column's data. For example, to access the 'claims' column:



````python
swedish_auto_zone_data['claims']
````


You can also use attribute-style access for column names that are valid Python identifiers and do not conflict with `DataFrame` methods:



````python
swedish_auto_zone_data.payments
````

You can also use dictionary-style assignment to modify or add new columns to a `DataFrame`. For instance, let's add a new column 'average severity' which is calculated as 'payments' divided by 'claims':



````python
swedish_auto_zone_data['average severity'] = swedish_auto_zone_data['payments'] / swedish_auto_zone_data['claims']
print(swedish_auto_zone_data)
````


We can now look at a `DataFrame` as an upgraded version of a two-dimensional array. Therefore, we can easily access its raw underlying data using the `values` attribute:



````python
swedish_auto_zone_data.values
````


Based on this point of view, we can therefore perform several array-operations on a `DataFrame`. For example, we can transpose a `DataFrame`:



````python
swedish_auto_zone_data.T
````


In particular, passing a single `index` to an array accesses a *row* (when we treat a `DataFrame` as a NumPy array):



````python
swedish_auto_zone_data.values[0]
````


On the other hand, passing a single `index` to a `DataFrame` accesses a *column*:



````python
swedish_auto_zone_data['payments']
````


## Selection and Filtering

Building upon basic indexing, Pandas offers powerful and flexible ways to select and filter data using boolean conditions, which are essential for targeted data analysis.

### Boolean Indexing (Masking)

Boolean indexing, also known as masking, uses a boolean Series or array to select data. Only rows (or columns) corresponding to `True` values in the boolean object are returned.



````python
# Re-using df_numeric from the Data Loading section
df_numeric = pd.DataFrame({
    'Age': [25, 30, 35, 28, 40, 55, 60, 45, 32, 29],
    'Premium': [1200, 1500, 1800, 1300, 2000, 2500, 2800, 2200, 1600, 1400],
    'Claims': [1, 0, 2, 0, 1, 3, 2, 0, 1, 0]
})

# Select rows where 'Age' is greater than 30
young_drivers = df_numeric[df_numeric['Age'] > 30]
print("Young drivers (Age > 30):\n", young_drivers)

# Select rows where 'Claims' is 0
no_claims = df_numeric[df_numeric['Claims'] == 0]
print("\nPolicies with no claims:\n", no_claims)
````


### Conditional Filtering with Multiple Criteria

You can combine multiple boolean conditions using logical operators (`&` for AND, `|` for OR, `~` for NOT). Each condition must be enclosed in parentheses.



````python
# Select policies where 'Age' is less than 40 AND 'Claims' is 1
specific_segment = df_numeric[(df_numeric['Age'] < 40) & (df_numeric['Claims'] == 1)]
print("Policies with Age < 40 AND Claims == 1:\n", specific_segment)

# Select policies where 'Premium' is greater than 2000 OR 'Claims' is greater than 2
high_value_or_risk = df_numeric[(df_numeric['Premium'] > 2000) | (df_numeric['Claims'] > 2)]
print("\nPolicies with Premium > 2000 OR Claims > 2:\n", high_value_or_risk)

# Select policies where 'Age' is NOT between 30 and 50 (inclusive)
not_mid_age = df_numeric[~((df_numeric['Age'] >= 30) & (df_numeric['Age'] <= 50))]
print("\nPolicies not in mid-age (30-50):\n", not_mid_age)
````


### The `isin()` Method

The `isin()` method is useful for filtering rows where a column's value is among a list of specified values.



````python
# Filter for specific Patient_IDs using df_patients
df_patients = pd.DataFrame({
    'Patient_ID': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    'Patient_SSN': ['8392206319', '8473303738', '7453409011', '5904533830', '8482778872', '5528263293', '2482671444'],
    'Doctor_ID': [109, 108, 109, 110, 101, 103, 104],
    'Patient_Name': ['John', 'Akash', 'Irfan', 'Monica', 'Charlie', 'Angelina', 'Tony'],
    'Gender': ['Male', 'Male', 'Male', 'Female', 'Male', 'Female', 'Male']
})

target_ids = [1001, 1003, 1005]
filtered_patients = df_patients[df_patients['Patient_ID'].isin(target_ids)]
print("Patients with specific IDs (1001, 1003, 1005):\n", filtered_patients)
````


## Operating on Data in Pandas

Pandas leverages NumPy's efficient element-wise operations (Universal Functions or ufuncs) for arithmetic, trigonometric, exponential, and logarithmic functions. However, Pandas adds two critical features to these operations:

1.  **Index Preservation:** When performing unary operations (e.g., negation, `sin()`), the index and column labels are preserved in the output, ensuring data context is maintained.
2.  **Index Alignment:** For binary operations (e.g., addition, subtraction between two Series or DataFrames), Pandas automatically aligns the indices. This means operations are performed on matching labels, and non-matching labels result in `NaN` (Not a Number) values.

### Ufuncs: Index Preservation

Any NumPy ufunc can be applied to Pandas `Series` and `DataFrames`, and the result will be another Pandas object with its indices preserved.



````python
np.random.seed(42) # For reproducibility
series_rand = pd.Series(np.random.randint(0, 10, 5), index=list('abcde'))
dataframe_rand = pd.DataFrame(np.random.randint(0, 10, (4, 3)), columns=list('XYZ'))
print("Original Series:\n", series_rand)
print("\nOriginal DataFrame:\n", dataframe_rand)
````


Applying ufuncs preserves the structure:



````python
print("Exponential of Series:\n", np.exp(series_rand))
print("\nSine of DataFrame (scaled by pi):\n", np.sin(dataframe_rand * np.pi))
````


### Ufuncs: Index Alignment

One of Pandas' most powerful features is automatic index alignment during binary operations. This ensures that operations are performed on corresponding data, even if the objects have different or unsorted indices.

Let's demonstrate with two `Series` that have partially overlapping and different indices:



````python
series_A = pd.Series([0, 1, 2], index=['a', 'b', 'c'])
series_B = pd.Series([3, 4, 5], index=['b', 'c', 'd'])
print("Series A:\n", series_A)
print("\nSeries B:\n", series_B)
````


When we add them, Pandas aligns the indices:



````python
print("Series A + Series B:\n", series_A + series_B)
````


The result contains the *union* of indices from both `Series`. For any index where a value is missing in either `Series`, the result is `NaN` (Not a Number), indicating missing data. This behavior is crucial for robust data analysis when combining datasets with potentially inconsistent indices.

You can control the fill value for missing entries during alignment using the corresponding object methods (e.g., `.add()`, `.subtract()`) instead of operators. For example, to fill `NaN` values with `0` during subtraction:



````python
print("Series A - Series B (fill with 0):\n", series_A.subtract(series_B, fill_value=0))
````


> **💡 TIP**
> Using `series_5 - series_6` is equivalent to `series_5.substract(series_6)`, however, the later allows us to specify the fill value for any missing values.
>

Similar alignment rules apply to `DataFrames`, where both row and column indices are aligned.



````python
np.random.seed(42)
df1 = pd.DataFrame(np.random.randint(0, 10, (3, 3)), columns=list('ADC'), index=list('xyz'))
df2 = pd.DataFrame(np.random.randint(0, 10, (4, 4)), columns=list('BDCE'), index=list('zyxw'))
print("DataFrame 1:\n", df1)
print("\nDataFrame 2:\n", df2)
````


Performing a subtraction results in alignment across both axes:



````python
print("DataFrame 1 - DataFrame 2:\n", df1 - df2)
````


Notice that `NaN` values appear where either DataFrame does not have a corresponding row or column label. The resulting indices are sorted. You can also specify a `fill_value` here, for instance, filling with the mean of `df1`:



````python
fill_value_df1 = df1.stack().mean()
print("DataFrame 1 - DataFrame 2 (fill with mean of df1):\n", df1.subtract(df2, fill_value=fill_value_df1))
````


The following table summarizes Python operators and their equivalent Pandas object methods:

| Python Operator | Pandas method(s) |
| --- | --- |
| `+`             | `add()`          |
| `-`             | `sub()`, `subtract()` |
| `*`             | `mul()`, `multiply()` |
| `/`             | `truediv()`, `div()`, `divide()` |
| `//`            | `floordiv()`     |
| `%`             | `mod()`          |
| `**`            | `pow()`          |

**Operations between a DataFrame and a Series (Broadcasting):**
Operations between a `DataFrame` and a `Series` are similar to NumPy's broadcasting rules for two-dimensional and one-dimensional arrays.

By default, operations between a `DataFrame` and a `Series` are performed *row-wise* (across columns).



````python
np.random.seed(42)
df_example = pd.DataFrame(np.random.randint(0, 5, (2, 3)), columns=list('abc'))
print("Original DataFrame:\n", df_example)
print("\nAdding first row (broadcasted row-wise):\n", df_example + df_example.iloc[0])
````


To operate *column-wise* (down rows), you need to explicitly specify the `axis` parameter in the method call. For instance, to add the 'c' column to each column of the DataFrame:



````python
print("Adding column 'c' (broadcasted column-wise with axis=0):\n", df_example.add(df_example['c'], axis=0))
````


Remember that index alignment still applies. If the `Series`'s index does not match the `DataFrame`'s index along the specified axis, `NaN` values will result.

## Grouping and Aggregation

`Groupby` is one of the most powerful and frequently used operations in data analysis. It allows you to group data based on one or more keys (columns), apply a function to each group independently, and then combine the results into a new DataFrame. This split-apply-combine strategy is fundamental for summarizing and analyzing data by categories.

### The Split-Apply-Combine Strategy

The `groupby()` operation involves three main steps:

1.  **Split:** The data is divided into groups based on some criterion (e.g., unique values in a column).
2.  **Apply:** A function (e.g., aggregation, transformation, or filtering) is applied to each group independently.
3.  **Combine:** The results from each group are combined into a single, cohesive data structure.

Let's start with a simple example using a DataFrame of sales data.



````python
# Create a sample DataFrame for demonstration
df_sales = pd.DataFrame({
    'Region': ['North', 'South', 'North', 'West', 'South', 'North', 'West'],
    'Salesperson': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Alice', 'Frank'],
    'Sales': [100, 150, 120, 200, 130, 110, 180],
    'Units': [10, 15, 12, 20, 13, 11, 18]
})
print("Original Sales DataFrame:\n", df_sales)
````


### Simple Aggregations

The most common use of `groupby()` is to perform aggregation functions, which compute a single summary statistic for each group. Common aggregation functions include `sum()`, `mean()`, `count()`, `min()`, `max()`, and `median()`.

#### Grouping by a Single Column

Let's calculate the total `Sales` for each `Region`:



````python
# Group by 'Region' and calculate the sum of 'Sales'
sales_by_region = df_sales.groupby('Region')['Sales'].sum()
print("Total Sales by Region:\n", sales_by_region)
````


Similarly, we can find the average `Units` sold by each `Salesperson`:



````python
# Group by 'Salesperson' and calculate the mean of 'Units'
average_units_by_salesperson = df_sales.groupby('Salesperson')['Units'].mean()
print("\nAverage Units by Salesperson:\n", average_units_by_salesperson)
````


#### Grouping by Multiple Columns

You can group by multiple columns by passing a list of column names to `groupby()`.



````python
# Group by 'Region' and 'Salesperson' and calculate total sales
sales_by_region_salesperson = df_sales.groupby(['Region', 'Salesperson'])['Sales'].sum()
print("Total Sales by Region and Salesperson:\n", sales_by_region_salesperson)

# The result is a Series with a MultiIndex. You can unstack it for a DataFrame view:
print("\nTotal Sales by Region and Salesperson (Unstacked):\n", sales_by_region_salesperson.unstack(fill_value=0))
````


#### Applying Multiple Aggregation Functions (`agg()`)

To apply multiple aggregation functions to one or more columns, use the `agg()` method. You can pass a list of function names (as strings) or a dictionary mapping column names to functions.



````python
# Apply multiple aggregations to 'Sales'
multi_agg_sales = df_sales.groupby('Region')['Sales'].agg(['sum', 'mean', 'count'])
print("Multiple Aggregations on Sales by Region:\n", multi_agg_sales)

# Apply different aggregations to different columns
custom_agg = df_sales.groupby('Region').agg(
    Total_Sales=('Sales', 'sum'),
    Average_Units=('Units', 'mean'),
    Num_Transactions=('Sales', 'count')
)
print("\nCustom Aggregations by Region:\n", custom_agg)
````


### Transformation

The `transform()` method returns an object that has the same index as the original DataFrame, filled with transformed values (e.g., values normalized within their group, or the group mean/sum broadcasted to all original rows). This is useful for feature engineering.



````python
# Add a new column showing each salesperson's sales as a percentage of their region's total sales
df_sales['Regional_Sales_Total'] = df_sales.groupby('Region')['Sales'].transform('sum')
df_sales['Sales_Share_of_Region'] = df_sales['Sales'] / df_sales['Regional_Sales_Total']
print("DataFrame with Regional Sales Total and Share:\n", df_sales)
````


### Filtering

The `filter()` method is used to keep or discard entire groups based on a condition applied to the group data. The condition must return `True` or `False` for each group.



````python
# Filter regions where the total sales are greater than 250
high_sales_regions = df_sales.groupby('Region').filter(lambda x: x['Sales'].sum() > 250)
print("\nRegions with Total Sales > 250:\n", high_sales_regions)
````


This method is particularly useful for actuarial tasks such as identifying high-risk segments (e.g., groups of policies with unusually high claim frequencies) or filtering out data points from underrepresented groups for further analysis. The `groupby()` operation is incredibly versatile and forms the backbone of many complex data analysis tasks in Pandas.

## Applying Functions

Pandas provides powerful methods like `apply()` and `map()` that enable you to perform custom operations on your data. These methods are essential for transforming data in ways that built-in functions might not directly support, such as applying complex business logic or reformatting specific columns.

### `apply()` for Series and DataFrames

The `apply()` method can be used on both Series and DataFrames. It passes each Series (for DataFrame columns or rows) or each element (for a Series) to a custom function.

#### Applying to a Series

When `apply()` is used on a Series, it applies the function to each element.



````python
# Create a sample Series
ages = pd.Series([25, 30, 35, 40, 45, 50, 55])
print("Original Ages Series:\n", ages)

# Define a custom function to categorize age
def categorize_age(age):
    if age < 30:
        return 'Young'
    elif 30 <= age <= 50:
        return 'Middle-aged'
    else:
        return 'Senior'

# Apply the custom function to the Series
age_categories = ages.apply(categorize_age)
print("\nAge Categories:\n", age_categories)
````


#### Applying to DataFrame Columns or Rows

When `apply()` is used on a DataFrame, by default (`axis=0`), it applies the function to each column (as a Series). If `axis=1`, it applies the function to each row.



````python
# Re-using df_policies from Pivot Tables section
df_policies = pd.DataFrame({
    'Policy_ID': ['P101', 'P102', 'P103', 'P104', 'P105', 'P106', 'P107', 'P108'],
    'Region': ['East', 'West', 'East', 'North', 'South', 'West', 'North', 'East'],
    'Product': ['Auto', 'Home', 'Auto', 'Life', 'Auto', 'Home', 'Life', 'Auto'],
    'Premium': [1200, 1800, 1500, 2500, 1300, 2000, 2200, 1400],
    'Claims': [1, 0, 1, 0, 0, 1, 0, 1],
    'Customer_Segment': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B']
})

# Define a function to calculate claim ratio
def calculate_claim_ratio(row):
    if row['Premium'] > 0:
        return row['Claims'] / row['Premium']
    return 0

# Apply the function row-wise to calculate claim ratio for each policy
df_policies['Claim_Ratio'] = df_policies.apply(calculate_claim_ratio, axis=1)
print("DataFrame with Claim Ratio:\n", df_policies)
````


### `map()` for Series

The `map()` method is specifically for Series and is used to substitute each value in a Series with another value. It is very useful for cleaning and standardizing data. It can take a dictionary or another Series for mapping values.

#### Mapping with a Dictionary



````python
# Map product codes to full names
product_mapping = {'Auto': 'Automobile Insurance', 'Home': 'Homeowner's Insurance', 'Life': 'Life Insurance'}

df_policies['Product_Full_Name'] = df_policies['Product'].map(product_mapping)
print("DataFrame with Product Full Name:\n", df_policies)
````


#### Mapping with a Function

You can also pass a function (including a `lambda` function) to `map()`.



````python
# Create a dummy Series for demonstration
policy_status = pd.Series(['Active', 'Lapsed', 'Active', 'Pending', 'Lapsed'])
print("Original Policy Status:\n", policy_status)

# Map status to a boolean indicating if it's an active policy
is_active = policy_status.map(lambda x: True if x == 'Active' else False)
print("\nIs Active Status:\n", is_active)
````


> **💡 TIP**
> -   Use `apply()` when you need to apply a function along an axis of a DataFrame (row-wise or column-wise) or to each element of a Series where the logic might be more complex and depend on other values in the row/column.
> -   Use `map()` when you have a Series and want to substitute values based on a dictionary, another Series, or a simple function applied element-wise.
>

## Time Series

Time series data is ubiquitous in actuarial science, from tracking claim frequencies and severity over time to analyzing mortality rates, premium trends, and investment performance. Pandas provides powerful and flexible tools for working with time-indexed data, making it straightforward to analyze, manipulate, and visualize temporal patterns.

### Converting to Datetime Objects

The first step in working with time series data in Pandas is often to convert date/time strings into proper `datetime` objects using `pd.to_datetime()`. This allows Pandas to recognize and leverage the temporal nature of the data.



````python
# Create a DataFrame with a date string column
df_time = pd.DataFrame({
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05'],
    'Value': [10, 12, 15, 11, 13]
})
print("Original DataFrame with date strings:\n", df_time)
print("Date column dtype before conversion:", df_time['Date'].dtype)

# Convert the 'Date' column to datetime objects
df_time['Date'] = pd.to_datetime(df_time['Date'])
print("\nDataFrame after date conversion:\n", df_time)
print("Date column dtype after conversion:", df_time['Date'].dtype)
````


### Setting Datetime as Index

For most time series operations, it's highly beneficial to set the `datetime` column as the DataFrame's index. This enables powerful time-based indexing and resampling capabilities.



````python
# Set the 'Date' column as the DataFrame index
df_time_indexed = df_time.set_index('Date')
print("DataFrame with Datetime Index:\n", df_time_indexed)
````


### Basic Time-Based Indexing and Slicing

With a `DatetimeIndex`, you can easily select data using dates and time ranges.



````python
# Select data for a specific date
print("Value on 2023-01-03:\n", df_time_indexed.loc['2023-01-03'])

# Select data for a range of dates (inclusive)
print("\nValues from 2023-01-02 to 2023-01-04:\n", df_time_indexed.loc['2023-01-02':'2023-01-04'])

# Select data by year or month (if index contains such granularity)
# Example: if you had data for multiple months, you could select df_time_indexed.loc['2023-01']
# For this small dataset, we'll just demonstrate partial string indexing
print("\nValues for January 2023 (partial string indexing):\n", df_time_indexed.loc['2023-01'])
````


### Resampling Time Series Data

Resampling is a crucial operation for changing the frequency of time series data (e.g., from daily to weekly, or monthly to quarterly). It involves binning data into new time intervals and then applying an aggregation function.



````python
# Create a more granular time series for resampling demonstration
# Daily data over a few weeks
idx = pd.to_datetime([f'2023-01-{d:02d}' for d in range(1, 22)]) # Jan 1 to Jan 21
values = np.random.randint(5, 20, len(idx))
df_daily = pd.DataFrame({'Claims_Count': values}, index=idx)
print("Original Daily Claims Count:\n", df_daily.head())

# Resample to weekly frequency, taking the sum of claims for each week
df_weekly_sum = df_daily.resample('W').sum()
print("\nWeekly Claims Count (Sum):\n", df_weekly_sum)

# Resample to weekly frequency, taking the mean of claims for each week
df_weekly_mean = df_daily.resample('W').mean()
print("\nWeekly Claims Count (Mean):\n", df_weekly_mean)

# Resample to monthly frequency, taking the sum
df_monthly_sum = df_daily.resample('M').sum()
print("\nMonthly Claims Count (Sum):\n", df_monthly_sum)
````


> **💡 TIP**
> Common resampling aliases:
>
> -   `B`: Business day frequency
> -   `D`: Calendar day frequency
> -   `W`: Weekly frequency
> -   `M`: Month end frequency
> -   `Q`: Quarter end frequency
> -   `A`: Year end frequency
>

Time series functionality in Pandas is a cornerstone for dynamic actuarial analysis, enabling everything from trend analysis and seasonality detection to forecasting and risk management over different time horizons.

## Handling Missing Data

Missing data is a common challenge in real-world datasets, especially in actuarial science where data might be incomplete or erroneous. Pandas provides robust tools to detect, remove, and replace these missing values.

> **📝 NOTE**
> We will use the terms `null`, `NaN` (Not a Number), or `NA` (Not Available) interchangeably to refer to missing data.
>

### Strategies for Representing Missing Data

There are two primary strategies for indicating missing data:

-   **Using a Mask:** An external Boolean array (or mask) explicitly indicates which values are missing.
-   **Using a Sentinel Value:** A specific, reserved value within the data itself (e.g., `NaN` for floating-point numbers) signifies missingness.

Each approach has trade-offs. Masks require additional storage and computation. Sentinel values, while integrated, can reduce the range of valid data that can be represented (e.g., if `0` or `-1` were used as sentinels). Pandas primarily uses sentinel values.

### Missing Data in Pandas (`NaN` and `None`)

Pandas intelligently handles `None` and `NaN` values, converting between them as necessary to maintain consistency and enable efficient operations. When `None` is introduced into a `Series` or `DataFrame` of a numeric type, Pandas will typically upcast the data type to a floating-point type to accommodate `NaN`.



````python
series_mixed = pd.Series([3, 8, np.nan, 1, None, 6])
print("Mixed Series:\n", series_mixed)
print("Data type:", series_mixed.dtype)
````


If you start with an integer `Series` and introduce `None`, it will be converted to `float64`:



````python
int_series = pd.Series(range(5), dtype=int)
print("Original Integer Series:\n", int_series)
int_series[2] = None # Assigning None to an integer Series
print("\nSeries after assigning None:\n", int_series)
print("Data type after None assignment:", int_series.dtype)
````


The following table summarizes how Pandas handles `NA` values across different data types:

| Typeclass | Conversion when storing NAs | NA Sentinel Value(s) |
| --- | --- | --- |
| `floating`| No change                   | `np.nan`             |
| `object`  | No change                   | `None` or `np.nan`   |
| `integer` | Cast to `float64`           | `np.nan`             |
| `boolean` | Cast to `object`            | `None` or `np.nan`   |

Let's illustrate these conversions:



````python
# Floating-point (no change)
float_series = pd.Series([1.0, 2.0, 3.0])
float_series[1] = np.nan
print("Float Series with NaN:\n", float_series)
print("Float Series dtype:", float_series.dtype)

# Object (no change)
object_series = pd.Series(['apple', 'banana', 'cherry'])
object_series[1] = None
print("\nObject Series with None:\n", object_series)
print("Object Series dtype:", object_series.dtype)

# Boolean (casts to object)
bool_series = pd.Series([True, False, True])
bool_series[1] = np.nan
print("\nBoolean Series with NaN:\n", bool_series)
print("Boolean Series dtype:", bool_series.dtype)
````


> **📝 NOTE**
> Strings in Pandas are always stored with an `object` data type.
>

### Operating on Null Values

Pandas provides several convenient methods for managing null values:

-   `isnull()`: Returns a Boolean mask indicating `True` where values are missing.
-   `notnull()`: Returns a Boolean mask indicating `True` where values are *not* missing.
-   `dropna()`: Returns a new `Series` or `DataFrame` with null values removed.
-   `fillna()`: Returns a new `Series` or `DataFrame` with null values replaced (imputed).

#### Detecting Null Values

The `isnull()` and `notnull()` methods are fundamental for identifying missing data. They both return a Boolean mask of the same shape as the original object.



````python
series_na = pd.Series([None, "Value A", 'Value B', np.nan, None, "Value C"])
print("Original Series:\n", series_na)
````


Using `isnull()`:



````python
print("\nIsNull mask:\n", series_na.isnull())
print("\nNull values from Series:\n", series_na[series_na.isnull()])
````


Using `notnull()`:



````python
print("\nNotNull mask:\n", series_na.notnull())
print("\nNon-null values from Series:\n", series_na[series_na.notnull()])
````


#### Dropping Null Values

The `dropna()` method is used to remove rows or columns containing null values. By default, it drops any row that contains *at least one* null value.



````python
print("Series after dropping NA:\n", series_na.dropna())
````


For `DataFrames`, `dropna()` offers more control. Let's create a DataFrame with some null values:



````python
df_na = pd.DataFrame([[np.nan, 40, 10, 91],
                      [300, np.nan, 67, 32],
                      [104, 12, np.nan, 365],
                      [1, 743, 2, 5]])
print("Original DataFrame:\n", df_na)
````


By default, `dropna()` drops rows with any null values:



````python
print("DataFrame after dropping rows with any NA:\n", df_na.dropna())
````


To drop columns with null values, set `axis=1` (or `axis='columns'`):



````python
print("DataFrame after dropping columns with any NA:\n", df_na.dropna(axis=1))
````


**Controlling `dropna()` Behavior:**
You can fine-tune `dropna()` using the `how` and `thresh` parameters:

-   `how='any'` (default): Drops a row/column if *any* `NaN` is present.
-   `how='all'`: Drops a row/column only if *all* values are `NaN`.

Let's add a fully null column to `df_na`:



````python
df_na[4] = np.nan
print("DataFrame with an all-NA column:\n", df_na)
print("\nDataFrame after dropping columns with all NA (how='all'):\n", df_na.dropna(axis='columns', how='all'))
````


-   `thresh`: Requires a minimum number of non-null values for a row/column to be kept.



````python
print("DataFrame after dropping rows with less than 4 non-NA values (thresh=4):\n", df_na.dropna(axis='rows', thresh=4))
````

In the output, rows with indices 0, 1, and 2 are dropped because they have fewer than 4 non-null values (they each have 3).

#### Filling Null Values

Instead of removing missing data, you can replace (impute) them using the `fillna()` method. This allows you to substitute `NaN` values with a specified constant, or values propagated from other parts of the data.

Let's use a `Series` with null values:



````python
series_fill = pd.Series([None, np.nan, 100, 50, None, 25, 10], index=list('ABCDEFG'))
print("Original Series:\n", series_fill)
````


-   **Fill with a scalar value (e.g., zero):**


````python
print("Series filled with 0:\n", series_fill.fillna(0))
````


-   **Forward-fill (`method='ffill'`):** Propagates the last valid observation forward to the next valid observation.


````python
print("Series with forward-fill:\n", series_fill.fillna(method='ffill'))
````


> **💡 TIP**
> If the first value(s) are `NaN` and there's no previous value to propagate, they will remain `NaN`.
>

-   **Back-fill (`method='bfill'`):** Propagates the next valid observation backward to the previous valid observation.


````python
print("Series with back-fill:\n", series_fill.fillna(method='bfill'))
````


For `DataFrames`, `fillna()` also allows specifying the `axis` for filling (e.g., `axis=1` for row-wise filling).



````python
print("DataFrame with forward-fill (axis=1):\n", df_na.fillna(method='ffill', axis=1))
````


## Hierarchical Indexing (MultiIndex)

When working with higher-dimensional data that is indexed by more than one key, *hierarchical indexing* (also known as *MultiIndex*) becomes invaluable. It allows you to incorporate multiple index levels within a single `Series` or `DataFrame` index, effectively representing data of three or more dimensions in a more compact two-dimensional structure.

This section will cover the creation of `MultiIndex` objects, performing indexing and slicing on multiply indexed data, and converting between simple and multi-level indexed objects.

To illustrate, let's consider a scenario from actuarial science where we track claims data for different geographic zones and car models. Suppose we have claims data for three geographic zones (`Zone = 1, 2, 3`) and nine car models (`Make = 1` through `9`). We can represent this two-dimensional data within a one-dimensional `Series` using a `MultiIndex`.

First, let's manually create a list of tuples to serve as our multi-index, where each tuple `(zone, make)` uniquely identifies an observation:



````python
multi_index_tuples = [('Zone1', 'Make1'), ('Zone1', 'Make2'), ('Zone1', 'Make3'), ('Zone1', 'Make4'), 
                      ('Zone1', 'Make5'), ('Zone1', 'Make6'), ('Zone1', 'Make7'), ('Zone1', 'Make8'), 
                      ('Zone1', 'Make9'),
                      ('Zone2', 'Make1'), ('Zone2', 'Make2'), ('Zone2', 'Make3'), ('Zone2', 'Make4'), 
                      ('Zone2', 'Make5'), ('Zone2', 'Make6'), ('Zone2', 'Make7'), ('Zone2', 'Make8'), 
                      ('Zone2', 'Make9'),
                      ('Zone3', 'Make1'), ('Zone3', 'Make2'), ('Zone3', 'Make3'), ('Zone3', 'Make4'), 
                      ('Zone3', 'Make5'), ('Zone3', 'Make6'), ('Zone3', 'Make7'), ('Zone3', 'Make8'), 
                      ('Zone3', 'Make9')]

claims_data = [2186, 607, 455, 504, 528, 652, 332, 142, 17768,
               2157, 617, 352, 421, 566, 724, 386, 140, 15939,
               1898, 430, 298, 392, 593, 885, 427, 189, 14826]

swedish_claims_multiindex = pd.Series(claims_data, index=multi_index_tuples)
print("Series with tuple index (not yet MultiIndex):\n", swedish_claims_multiindex)
````


While the above Series has a tuple as its index, it's not yet a true Pandas `MultiIndex`. To convert it, we can use `pd.MultiIndex.from_tuples()`:



````python
multi_index = pd.MultiIndex.from_tuples(multi_index_tuples, names=['Zone', 'Make'])
print("Created MultiIndex:\n", multi_index)
````


Now, let's reindex our Series with this `MultiIndex`. Pandas will display a hierarchical representation of the data, which is much more readable and intuitive:



````python
swedish_claims_multiindex = swedish_claims_multiindex.reindex(multi_index)
print("Series with MultiIndex (hierarchical representation):\n", swedish_claims_multiindex)
````


> **📝 NOTE**
> In the hierarchical display, blank entries in an outer index level indicate that the value is the same as the non-blank entry directly above it.
>

**MultiIndex Indexing and Slicing:** The advantage of a `MultiIndex` becomes apparent when you need to select data based on multiple levels. You can use direct tuple indexing for specific entries or partial indexing for entire blocks.

For example, to select all claims for 'Make8' across all zones:



````python
print("Claims for Make8:\n", swedish_claims_multiindex.loc[:, 'Make8'])
````


To select claims for 'Zone2' specifically:



````python
print("Claims for Zone2:\n", swedish_claims_multiindex.loc['Zone2'])
````


And to select claims for 'Zone1' from 'Make3' to 'Make8':



````python
print("Claims for Zone1, Make3 to Make8:\n", swedish_claims_multiindex.loc[('Zone1', 'Make3'):('Zone1', 'Make8')])
````


**Converting between MultiIndex and DataFrame:**
The `unstack()` method allows you to convert a `Series` with a `MultiIndex` into a `DataFrame` with a single index and multi-level columns. This is essentially pivoting the inner index level into columns.



````python
swedish_claims_df = swedish_claims_multiindex.unstack()
print("DataFrame (unstacked Make levels into columns):\n", swedish_claims_df)
````


Conversely, the `stack()` method performs the opposite operation, converting columns (or a subset of columns) into an inner index level, returning a `Series` (or a `DataFrame` with fewer columns):



````python
print("Series (stacked back from DataFrame):\n", swedish_claims_df.stack())
````


**Representing Higher-Dimensional Data:**
The true power of `MultiIndex` is in representing data with more than two dimensions. Each additional level in the `MultiIndex` effectively adds another dimension to your data. For example, if we also have payment data corresponding to each `(Zone, Make)` pair:



````python
payments_data = [32257.56, 7326.41, 8120.98, 9701.43, 6518.68, 13342.66, 5806.94, 1754.56, 241564.88,
                 37740.52, 8894.96, 7575.48, 11549.92, 8174.16, 18156.65, 8046.79, 2746.83, 285031.47,
                 41609.79, 8708.93, 8336.88, 12689.94, 10053.51, 23627.86, 9254.79, 4084.29, 310966]

# Create a DataFrame with both claims and payments, using the MultiIndex
swedish_auto_multi_df = pd.DataFrame({'claims': swedish_claims_multiindex, 'payments': payments_data})
print("DataFrame with MultiIndex and multiple data columns:\n", swedish_auto_multi_df)
````


With this structure, we can easily perform operations like calculating the average severity (payments per claim) for each `(Zone, Make)` combination, and then unstack it for a clear tabular view:



````python
average_severity_multiindex = swedish_auto_multi_df['payments'] / swedish_auto_multi_df['claims']
print("Average Severity (unstacked):\n", average_severity_multiindex.unstack())
````


## Concatenation

Combining different datasets is a common task in data analysis, ranging from simple concatenation to more complex database-style joins and merges. Pandas provides powerful functions and methods for these manipulations.

We will first explore simple concatenation of `Series` and `DataFrames` using `pd.concat()`, and then delve into more complex join and merge operations.

**Recall: Concatenation of NumPy Arrays**
As a refresher, recall how NumPy's `np.concatenate()` works for combining arrays:



````python
array_a = np.array([1, 2, 3])
array_b = np.array([4, 5, 6])
array_c = np.array([[7, 8], [9, 10]])
````

````python
print("Concatenating 1D arrays:\n", np.concatenate([array_a, array_b]))
print("\nConcatenating 2D arrays along axis 0:\n", np.concatenate([array_c, array_c], axis=0))
print("\nConcatenating 2D arrays along axis 1:\n", np.concatenate([array_c, array_c], axis=1))
````

The `np.concatenate()` function takes a list or tuple of arrays as its first argument and an optional `axis` keyword to specify the dimension along which to concatenate.

Pandas `pd.concat()` is highly versatile and similar to `np.concatenate()`, but with additional features designed for handling Pandas objects' indices. Its full documentation can be found [here](https://pandas.pydata.org/docs/reference/api/pandas.concat.html).

**Concatenating Series:**


````python
series_concat_1 = pd.Series(['A', 'B', 'C'], index=[1, 2, 3])
series_concat_2 = pd.Series(['D', 'E', 'F'], index=[4, 5, 6])
print("Concatenated Series:\n", pd.concat([series_concat_1, series_concat_2]))
````


**Concatenating DataFrames (Row-wise):**
By default, `pd.concat()` concatenates DataFrames row-wise (`axis=0`).



````python
df_concat_1 = pd.DataFrame({'col1': ['A0', 'A1'], 'col2': ['B0', 'B1']}, index=[0, 1])
df_concat_2 = pd.DataFrame({'col1': ['A2', 'A3'], 'col2': ['B2', 'B3']}, index=[2, 3])
print("DataFrame 1:\n", df_concat_1)
print("\nDataFrame 2:\n", df_concat_2)
print("\nConcatenated DataFrames (row-wise):\n", pd.concat([df_concat_1, df_concat_2]))
````


**Concatenating DataFrames (Column-wise):**
To concatenate column-wise, set `axis=1`.



````python
df_concat_3 = pd.DataFrame({'col_A': ['A0', 'A1'], 'col_B': ['B0', 'B1']}, index=[0, 1])
df_concat_4 = pd.DataFrame({'col_C': ['C0', 'C1'], 'col_D': ['D0', 'D1']}, index=[0, 1])
print("DataFrame 3:\n", df_concat_3)
print("\nDataFrame 4:\n", df_concat_4)
print("\nConcatenated DataFrames (column-wise):\n", pd.concat([df_concat_3, df_concat_4], axis=1))
````


**Handling Duplicate Indices:**
Unlike `np.concatenate()`, `pd.concat()` preserves indices even if it results in duplicates. This can be either intentional or a potential issue. Pandas provides ways to manage this:



````python
df_dup_idx_1 = pd.DataFrame({'data': ['A', 'B']}, index=[1, 2])
df_dup_idx_2 = pd.DataFrame({'data': ['C', 'D']}, index=[1, 2]) # Duplicate indices
print("DataFrame with duplicate indices (before concat):\n", pd.concat([df_dup_idx_1, df_dup_idx_2]))
````


-   **Catching duplicates as an error:** Use `verify_integrity=True` to raise a `ValueError` if duplicate indices are found.


````python
try:
    pd.concat([df_dup_idx_1, df_dup_idx_2], verify_integrity=True)
except ValueError as e:
    print(f"ValueError: {e}")
````


-   **Ignoring the index:** If the original indices are not important, you can generate a new, clean integer index using `ignore_index=True`.


````python
print("Concatenated with ignored index:\n", pd.concat([df_dup_idx_1, df_dup_idx_2], ignore_index=True))
````


-   **Adding MultiIndex Keys:** You can create a hierarchically indexed (MultiIndex) result by providing a `keys` argument. This adds an additional level to the index, making each original DataFrame's index distinct under a new key.


````python
print("Concatenated with MultiIndex keys:\n", pd.concat([df_dup_idx_1, df_dup_idx_2], keys=['dfA', 'dfB']))
````


**Concatenating DataFrames with Different Column Names:**
When concatenating DataFrames with non-matching column names, `pd.concat()` by default performs an 'outer' join, meaning it takes the union of all column names and fills missing entries with `NaN`.



````python
df_cols_1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df_cols_2 = pd.DataFrame({'B': [5, 6], 'C': [7, 8]})
print("DataFrame with different columns (outer join by default):\n", pd.concat([df_cols_1, df_cols_2]))
````


Alternatively, you can specify an 'inner' join using `join='inner', which will only include columns that are common to all input DataFrames (the intersection of columns).



````python
print("DataFrame with different columns (inner join):\n", pd.concat([df_cols_1, df_cols_2], join='inner'))
````


## Merge and Join

Pandas' `pd.merge()` function and the associated `.join()` method are powerful tools for combining `Series` or `DataFrame` objects based on common columns or indices, similar to SQL-style joins. This allows for rich data integration capabilities.

`pd.merge()` can perform various types of joins, automatically inferring the type based on the input data or allowing explicit specification:

#### One-to-One Joins

A one-to-one join combines two DataFrames where each key value in the 'left' DataFrame matches at most one key value in the 'right' DataFrame. This is akin to column-wise concatenation but based on a shared key.

Let's consider two DataFrames containing (fictitious) patient and social insurance number information:



````python
df_social_insurance = pd.DataFrame({
    'Social_Insurance_Number': ['SNQ657567', 'STS474766', 'SIS970766', 'SMS990766', 'SOS940766', 'SPL987866', 'RGI000234'],
    'Patient_ID': [1003, 1001, 1002, 1006, 1007, 1004, 1005]
})

print("Patients DataFrame:\n", df_patients)
print("\nSocial Insurance DataFrame:\n", df_social_insurance)
````


To combine these, we use `pd.merge()`. Pandas automatically detects the common column, 'Patient_ID', and uses it as the merge key:



````python
merged_patient_info = pd.merge(df_patients, df_social_insurance)
print("Merged Patient Information:\n", merged_patient_info)
````


**Key points about `pd.merge()`:**
-   It automatically identifies common column names to use as merge keys. You can explicitly specify the key(s) using the `on`, `left_on`, or `right_on` arguments.
-   The order of rows in the result is not necessarily preserved from the original DataFrames.
-   By default, `pd.merge()` discards the original indices of the input DataFrames and creates a new default integer index for the merged result.

#### Many-to-One Joins

In a many-to-one join, one of the key columns contains duplicate entries, while the other is unique. The resulting DataFrame will duplicate entries from the 'one' side to match all corresponding entries on the 'many' side.

Let's merge the `df_patients` DataFrame with a DataFrame containing doctor information, where multiple patients can be associated with one doctor:



````python
df_doctors = pd.DataFrame({
    'Doctor_ID': [101, 103, 104, 108, 109, 110],
    'Doctor_SSN': ['1266071613', '9017595693', '9190841246', '4931814876', '7453409011', '8888888888'],
    'Doctor_Name': ['Akshay', 'Hua', 'John', 'Joe', 'Elizabeth', 'Dr. X'],
    'Phone_Number': [7407535102, 1694362521, 7630399848, 2291855091, 8795129630, 1112223333]
})

print("Patients DataFrame (again):\n", df_patients)
print("\nDoctors DataFrame:\n", df_doctors)
````


Now, merge these two DataFrames on 'Doctor_ID':



````python
merged_patient_doctor = pd.merge(df_patients, df_doctors)
print("Merged Patient-Doctor Information (Many-to-One):\n", merged_patient_doctor)
````

Notice how 'Dr. Elizabeth' (Doctor_ID 109) appears twice, as two patients are associated with her. Similarly for 'Dr. John' (Doctor_ID 104) and 'Dr. X' (Doctor_ID 110).

#### Many-to-Many Joins

If the key column in both the left and right array contains duplicates, then the result is a many-to-many join. The result will be a DataFrame where all possible pairings of matching keys are included, potentially leading to a much larger DataFrame.

Let's merge the `df_patients` DataFrame with a DataFrame representing paychecks, where a doctor might have multiple paychecks and each paycheck is associated with a doctor ID (which itself might be duplicated for patients):



````python
df_paychecks = pd.DataFrame({
    'Check_Number': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110, 111, 112, 113, 114, 115, 116, 117, 118],
    'Doctor_ID': [101, 102, 110, 101, 103, 104, 105, 107, 109, 104, 106, 107, 108, 110, 102, 104, 109, 101],
    'Salary': [11934.43, 7839.88, 6180.097, 8013.878, 6443.145, 7563.87, 10216.875, 7640.142, 7736.25, 8831.368, 7495.05, 7804.575, 10421.271, 10114.757, 7204.1, 9142.002, 9312.028, 6244.0585],
    'Pay_Date': ['2020-07-31', '2020-08-31', '2020-08-31', '2020-06-30', '2020-05-31', '2020-07-31', '2020-07-31', '2020-06-30', '2020-05-31', '2020-09-30', '2020-07-31', '2020-05-31', '2020-06-30', '2020-09-30', '2020-09-30', '2020-08-31', '2020-07-31', '2020-08-31']
})

print("Paychecks DataFrame:\n", df_paychecks)
````


Merging `df_patients` and `df_paychecks` on 'Doctor_ID' will create a many-to-many relationship:



````python
merged_patient_paycheck = pd.merge(df_patients, df_paychecks)
print("Merged Patient-Paycheck Information (Many-to-Many):\n", merged_patient_paycheck)
````


From this merged DataFrame, we can aggregate data, such as the total salary recorded for each doctor:



````python
print("Total Salary per Doctor:\n", merged_patient_paycheck.groupby('Doctor_ID')['Salary'].sum())
````



## Optional: Pivot Tables

Pivot tables are a powerful feature in Pandas, similar to those found in spreadsheet software, that allow you to reshape and summarize DataFrames. They are invaluable for exploring relationships between multiple variables and for creating aggregated views of your data. The `pivot_table()` function is highly flexible, supporting various aggregation functions and multiple levels of indexing for both rows and columns.

### Introduction to `pivot_table()`

The `pd.pivot_table()` function takes several key arguments:

-   `data`: The DataFrame to be pivoted.
-   `values`: The column(s) to aggregate.
-   `index`: The column(s) to use as row index(es).
-   `columns`: The column(s) to use as column index(es).
-   `aggfunc`: The aggregation function(s) to apply (e.g., `'mean'`, `'sum'`, `np.sum`).
-   `fill_value`: Value to replace missing values in the pivot table.

Let's use an example with insurance policy data:



````python
# Create a sample DataFrame for pivot table demonstration
df_policies = pd.DataFrame({
    'Policy_ID': ['P101', 'P102', 'P103', 'P104', 'P105', 'P106', 'P107', 'P108'],
    'Region': ['East', 'West', 'East', 'North', 'South', 'West', 'North', 'East'],
    'Product': ['Auto', 'Home', 'Auto', 'Life', 'Auto', 'Home', 'Life', 'Auto'],
    'Premium': [1200, 1800, 1500, 2500, 1300, 2000, 2200, 1400],
    'Claims': [1, 0, 1, 0, 0, 1, 0, 1],
    'Customer_Segment': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B']
})
print("Original Policies DataFrame:\n", df_policies)
````


### Basic Pivot Table

Let's calculate the average `Premium` by `Region` and `Product`:



````python
# Calculate the average premium by region and product
pivot_avg_premium = pd.pivot_table(df_policies, 
                                   values='Premium', 
                                   index='Region', 
                                   columns='Product', 
                                   aggfunc='mean')
print("Average Premium by Region and Product:\n", pivot_avg_premium)
````


Notice how `NaN` values appear where there are no corresponding `(Region, Product)` combinations. We can fill these with a `fill_value`.



````python
# Fill NaN values with 0
pivot_avg_premium_filled = pd.pivot_table(df_policies, 
                                        values='Premium', 
                                        index='Region', 
                                        columns='Product', 
                                        aggfunc='mean', 
                                        fill_value=0)
print("\nAverage Premium (NaN filled with 0):\n", pivot_avg_premium_filled)
````


### Multiple Aggregation Functions

You can specify multiple aggregation functions by passing a list to `aggfunc`.



````python
# Calculate sum and mean of Premium by Region and Product
pivot_multi_agg = pd.pivot_table(df_policies, 
                                 values='Premium', 
                                 index='Region', 
                                 columns='Product', 
                                 aggfunc=['sum', 'mean'], 
                                 fill_value=0)
print("\nSum and Mean Premium by Region and Product:\n", pivot_multi_agg)
````


### Multiple Indexing for Rows and Columns

You can create more complex pivot tables by using multiple columns for `index` and `columns`.



````python
# Pivot with multiple row and column indices
pivot_complex = pd.pivot_table(df_policies, 
                               values='Premium', 
                               index=['Region', 'Customer_Segment'], 
                               columns=['Product', 'Claims'], 
                               aggfunc='mean', 
                               fill_value=0)
print("\nComplex Pivot Table (multiple indices):\n", pivot_complex)
````


Pivot tables are incredibly versatile for summarizing large datasets and are a cornerstone of exploratory data analysis in actuarial science, allowing for quick insights into aggregated policy performance, claim distributions, and product profitability across different segments.
