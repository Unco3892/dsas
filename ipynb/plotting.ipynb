## Univariate vs. Multivariate Data

When working with data, it is important to understand the distinction between **univariate** and **multivariate** data, as this influences the type of visualizations you will use.

### Univariate Data

Univariate data involves a single variable. The goal is often to understand the distribution, central tendency, and spread of the data. Common visualizations for univariate data include:

- **Histograms**: To show the frequency distribution of the data.
- **Box Plots**: To visualize the spread and identify outliers.
- **Violin Plots**: To combine distribution and density information.

> **💡 TIP**
> #### Why Use Univariate Visualizations?
> Univariate visualizations are simpler and are often used as a first step in data exploration.
>

### Multivariate Data

Multivariate data involves two or more variables. The goal is to understand relationships, correlations, and patterns between variables. Common visualizations for multivariate data include:

- **Scatter Plots**: To show relationships between two variables.
- **Pair Plots**: To explore relationships across multiple variables.
- **Heatmaps**: To visualize correlations in a matrix format.

> **💡 TIP**
> #### Why Use Multivariate Visualizations?
> Multivariate visualizations are crucial for identifying patterns and relationships that are not apparent in univariate analysis.
>

In this course, you will learn how to create both univariate and multivariate visualizations using Python libraries like Matplotlib and Seaborn. 

## Matplotlib Basics

Matplotlib is a versatile library for creating static, animated, and interactive visualizations in Python. It is often used for basic plotting tasks and provides a foundation for more advanced libraries like Seaborn.

### Anatomy of a Matplotlib Plot

A Matplotlib plot consists of several components:

- **Figure**: The overall container for the plot.
- **Axes**: The area where data is plotted (can have multiple axes in a figure).
- **Title**: Describes the plot.
- **Labels**: Text for the x-axis and y-axis.
- **Legend**: Explains the meaning of different elements in the plot.

### Importing Matplotlib

To use Matplotlib, you need to import it first:



```python
import matplotlib.pyplot as plt
```


### Creating a Simple Plot

Here is an example of a simple line plot:



```python
# Example: Line Plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```


This code creates a line plot with labeled axes and a title. The `plt.show()` function is used to display the plot.

### Customizing Plots

You can customize your plots by changing colors, line styles, and adding legends:



```python
# Example: Customized Line Plot
plt.plot(x, y, color='red', linestyle='--', marker='o', label='Line 1')
plt.title("Customized Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```


- **Color**: Changes the line color (e.g., 'red', 'blue').
- **Linestyle**: Changes the style of the line (e.g., '--', '-').
- **Marker**: Adds markers to data points (e.g., 'o', 's').
- **Grid**: Adds a grid to the plot for better readability.

### Subplots

Subplots allow you to create multiple plots in a single figure. This is useful for comparing different datasets or visualizations:



```python
# Example: Subplots
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st plot
plt.plot(x, y, 'r--')
plt.title("Plot 1")

plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd plot
plt.bar(x, y)
plt.title("Plot 2")

plt.tight_layout()
plt.show()
```


The `plt.subplot()` function specifies the layout of subplots. The `tight_layout()` function adjusts spacing to prevent overlap.

### Common Plot Types in Matplotlib

1. **Line Plot**: Used for trends over time.
2. **Bar Plot**: Used for categorical data.
3. **Histogram**: Used for frequency distributions.
4. **Scatter Plot**: Used for relationships between two variables.

## Seaborn Basics

Seaborn is built on top of Matplotlib and provides a high-level interface for creating attractive and informative statistical graphics. It simplifies the process of creating complex visualizations.

### Importing Seaborn

Normally, you do not need to install Seaborn manually, as it is often included in many Python environments. However, because we are using this specific interactive service for our lectures, we need to install it first.



```python
import micropip
micropip.install("seaborn")
```


Then, to use Seaborn, you need to import it:



```python
import seaborn as sns
```


### Creating a Simple Plot

Seaborn automatically applies a more visually appealing style to plots compared to Matplotlib.

Here is an example of a scatter plot:



```python
# Example: Scatter Plot
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)

sns.scatterplot(x=x, y=y)
plt.title("Simple Scatter Plot")
plt.show()
```


### Visualizing Distributions

Seaborn makes it easy to visualize distributions:



```python
# Example: Histogram
import numpy as np

data = np.random.randn(1000)

sns.histplot(data, kde=True)
plt.title("Histogram with KDE")
plt.show()
```


> **📝 NOTE**
> - **Histogram**: Shows the frequency of data points.
> - **KDE (Kernel Density Estimate)**: Smooths the histogram to show the probability density.
>

### Pair Plots

Pair plots are useful for exploring relationships between multiple variables:



```python
# Example: Pair Plot
import pandas as pd

data = pd.DataFrame({
    'A': np.random.rand(50),
    'B': np.random.rand(50),
    'C': np.random.rand(50)
})

sns.pairplot(data)
plt.show()
```


Pair plots create scatter plots for all variable combinations and histograms for individual variables.

### Heatmaps

Heatmaps are used to visualize correlations or matrices:



```python
# Example: Heatmap
correlation_matrix = data.corr()

sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```


> **📝 NOTE**
> - **annot=True**: Displays correlation values on the heatmap.
> - **cmap**: Changes the color scheme (e.g., 'coolwarm', 'viridis').
>

### Box Plots

Box plots are used to visualize the distribution of data and identify outliers:



```python
# Example: Box Plot
categories = ['A', 'B', 'C']
data = [
    np.random.rand(50),
    np.random.rand(50),
    np.random.rand(50)
]

sns.boxplot(data=data)
plt.xticks(ticks=[0, 1, 2], labels=categories)
plt.title("Box Plot Example")
plt.show()
```


### Violin Plots

Violin plots combine aspects of box plots and KDE plots to show distributions. Violin plots are ideal for visualizing the distribution and density of data across categories.



```python
# Example: Violin Plot
sns.violinplot(data=data)
plt.xticks(ticks=[0, 1, 2], labels=categories)
plt.title("Violin Plot Example")
plt.show()
```


### Advanced Customizations

Seaborn allows for advanced customizations to make your plots more informative and visually appealing:

> **📝 NOTE**
> #### Note
> You can customize Seaborn plots by combining them with Matplotlib functions for fine-grained control.
>



```python
# Example: Customized Scatter Plot
sns.scatterplot(x=x, y=y, hue=x > 0.5, style=x > 0.5, palette="coolwarm")
plt.title("Customized Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(title="Category")
plt.grid(True)
plt.show()
```


> **💡 TIP**
> ## Best Practices for Plotting
> 1. **Choose the Right Plot**: Match the plot type to the data and the story you want to tell.
> 2. **Label Everything**: Always include titles, axis labels, and legends.
> 3. **Keep It Simple**: Avoid clutter and focus on the key message.
> 4. **Use Colors Wisely**: Ensure colors are distinguishable and accessible.
> 5. **Test with Your Audience**: Make sure your visualizations are clear to others.
>
