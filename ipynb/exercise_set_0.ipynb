```python
# Install a package

pip uninstall package_name

pip list

pip show package_name
```

### Verifying Installations

After installing packages, verify they work correctly:



````python
# Comprehensive package verification
import importlib
import sys

# Define packages to check
packages = {
    'numpy': 'np',
    'pandas': 'pd',
    'matplotlib': 'mpl',
    'seaborn': 'sns',
    'scipy': 'scipy',
    'geopandas': 'gpd',
}

print("Package Installation Status")
print("=" * 50)

for package, alias in packages.items():
    try:
        # Import the package
        module = importlib.import_module(package)
        version = getattr(module, '__version__', 'Unknown')
        print(f"✓ {package:<15} {version:<10} OK")
    except ImportError as e:
        print(f"✗ {package:<15} {'---':<10} NOT INSTALLED")
        print(f"  Install with: pip install {package}")

# Test basic functionality
print("\nFunctionality Tests")
print("=" * 50)

try:
    import numpy as np
    arr = np.array([1, 2, 3, 4, 5])
    print(f"✓ NumPy array creation: {arr}")
except Exception as e:
    print(f"✗ NumPy test failed: {e}")

try:
    import pandas as pd
    df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
    print(f"✓ Pandas DataFrame shape: {df.shape}")
except Exception as e:
    print(f"✗ Pandas test failed: {e}")

try:
    import matplotlib.pyplot as plt
    fig, ax = plt.subplots()
    print(f"✓ Matplotlib figure created")
    plt.close(fig)
except Exception as e:
    print(f"✗ Matplotlib test failed: {e}")

try:
    import geopandas as gpd
    print(f"✓ Geopandas imported")
except Exception as e:
    print(f"✗ Geopandas test failed: {e}")

````

### Common Installation Issues and Solutions

#### Issue: "pip is not recognized"
**Solution**: 


```python
# Windows: Use Python to run pip
python -m pip install package_name

# Or add Python Scripts to PATH
# Usually: C:\Users\YourName\Anaconda3\Scripts
```

#### Issue: "Permission denied" errors
**Solution**:


```python
# Install for current user only
pip install --user package_name

# Or use conda instead
conda install package_name
```

#### Issue: Package conflicts
**Solution**:


```python
# Create a fresh environment
conda create -n fresh_env python=3.11
conda activate fresh_env

# Install packages one by one to identify conflicts
pip install numpy
pip install pandas
pip install geopandas
# etc.
```

#### Issue: "SSL Certificate" errors
**Solution**:


```python
# Temporarily trust PyPI
pip install --trusted-host pypi.org --trusted-host files.pythonhosted.org package_name

# Or use conda-forge channel
conda install -c conda-forge package_name
```

## Setting Up Your First Project

Let's create a structured workspace for the course:



````python
# Create a project structure for the course (run this in a Jupyter notebook or as a script)
import os

# Define project structure
project_name = "dsas_25"
folders = [
    "lectures",
    "labs",
    "data",
    "project",
    "scripts",
    "auxiliary",
]

# Create base directory
if not os.path.exists(project_name):
    os.makedirs(project_name)
    print(f"Created project folder: {project_name}")

# Create subdirectories
for folder in folders:
    path = os.path.join(project_name, folder)
    if not os.path.exists(path):
        os.makedirs(path)
        print(f"Created: {path}")

print("\nProject structure created successfully!")
````

## Your First Data Science Computation

Let's end with a practical actuarial calculation to test your setup:



````python
# Calculate the present value of an annuity
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Annuity parameters
annual_payment = 10000  # €10,000 per year
interest_rate = 0.035   # 3.5% annual rate  
years = 20              # 20-year annuity

# Calculate present value of each payment
time_periods = np.arange(1, years + 1)
discount_factors = (1 + interest_rate) ** -time_periods
present_values = annual_payment * discount_factors

# Create a DataFrame for clear presentation
annuity_df = pd.DataFrame({
    'Year': time_periods,
    'Payment': annual_payment,
    'Discount Factor': discount_factors,
    'Present Value': present_values
})

# Calculate total present value
total_pv = present_values.sum()

print(f"Annuity Analysis")
print(f"="*40)
print(f"Annual Payment: €{annual_payment:,}")
print(f"Interest Rate: {interest_rate:.1%}")
print(f"Term: {years} years")
print(f"Total Present Value: €{total_pv:,.2f}")
print(f"\nFirst 5 years:")
print(annuity_df.head())

# Visualize the present values
plt.figure(figsize=(10, 6))
plt.bar(time_periods, present_values, color='steelblue', alpha=0.7)
plt.xlabel('Year')
plt.ylabel('Present Value (€)')
plt.title('Present Value of Annual Payments')
plt.grid(axis='y', alpha=0.3)
plt.show()
````
